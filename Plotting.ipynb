Imports

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
```

Read Data

```python
df = pd.read_csv('data/processed/EDA3.csv', index_col=0)
df.head()
```

```python
counts = Counter(list(df['processed_Title']))
df_Title_Analysis = pd.DataFrame()
df_Title_Analysis['Words'] = counts.keys()
df_Title_Analysis['Frequency'] = counts.values()
df_Title_Analysis.to_csv('Data/processed/Title_Analysis.csv')
```

```python
plt.bar(*zip(*counts.items()))
plt.show()
```
